```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

input_dir = 'bird_sounds/'
output_dir = 'spectrograms/'

# Create output directory if it does not exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Set parameters for spectrogram
n_fft = 2048
hop_length = 512
n_mels = 128

# Recursively search for mp3 files and create spectrograms
for root, dirs, files in os.walk(input_dir):
    for file in files:
        if file.endswith('.mp3'):
            print(f"Processing file: {file}")
            
            # Load audio file
            file_path = os.path.join(root, file)
            y, sr = librosa.load(file_path)

            # Create spectrogram
    
            S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
           # S = librosa.feature.melspectrogram(y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
            S_DB = librosa.power_to_db(S, ref=np.max)

            # Save spectrogram as an image file
            
#             print(f"Saving spectrogram for {species} in {output_species_dir}")
            species = os.path.basename(os.path.dirname(root))

            species_dir = os.path.join(output_dir, species)
            if not os.path.exists(species_dir):
                os.makedirs(species_dir)
            output_path = os.path.join(species_dir, file[:-4] + '.png')
            librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length)
            plt.savefig(output_path)
            plt.clf()
            
print('Spectrograms generated successfully!')
```

```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

input_dir = 'spectrograms/'
output_dir = 'spectrograms_normalized/'

# Create output directory if it does not exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Set parameters for spectrogram
n_fft = 2048
hop_length = 512
n_mels = 128

# Normalize each spectrogram in the input directory
for root, dirs, files in os.walk(input_dir):
    for file in files:
        if file.endswith('.png'):
            print(f"Processing file: {file}")
            
            # Load spectrogram file
            file_path = os.path.join(root, file)
            S_DB = plt.imread(file_path)
            
            # Normalize the spectrogram by dividing each pixel value by the maximum pixel value
            S_DB_normalized = S_DB / np.max(S_DB)

            # Save the normalized spectrogram as a new image file
            species = os.path.basename(root)
            output_species_dir = os.path.join(output_dir, species)
            if not os.path.exists(output_species_dir):
                os.makedirs(output_species_dir)
            output_path = os.path.join(output_species_dir, file)
            plt.imsave(output_path, S_DB_normalized)
            
print('Spectrograms normalized successfully!')
```
